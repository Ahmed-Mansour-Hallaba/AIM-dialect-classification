```python
i=1
for i in range(0,len(dialect_df),1000):
#     print(i*1000,min((i+1)*1000,len(dialect_df)))
    print(i,i+1000)
#     ids=dialect_df['id'][i*1000:min((i+1)*1000,len(dialect_df))]
    ids=dialect_df['id'][i:min(i+1000,len(dialect_df))]
    ids_lists = ids.tolist()
    ids_stringlist = list(map(str, ids_lists))


    url = 'https://recruitment.aimtechnologies.co/ai-tasks'
    x = requests.post(url, json = ids_stringlist)
    data  = json.loads(x.text)

    content_df=pd.DataFrame.from_dict(data,orient='index',columns=['content']).reset_index()
    content_df['index']=content_df['index'].astype('int64')
    merged_df=dialect_df.merge(content_df, left_on='id', right_on='index').drop(columns=['id','index'])
    final_df=final_df.append(merged_df)
    if((i+1000)>len(dialect_df)):
        break
    i+=1
    
        
```
